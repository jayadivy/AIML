```python
# import the library opencv
import cv2
# globbing utility.
import glob
import matplotlib.pyplot as plt

# select the path
path = "image/*.*"
for file in glob.glob(path):
    image_read = cv2.imread(file)
  # conversion numpy array into rgb image to show
    c = cv2.cvtColor(image_read, cv2.COLOR_BGR2RGB)
    plt.imshow(c)
    plt.show()
```
